### Definitions

```python
def list_files(directory):
    try:
        # Get list of all files in the directory
        files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
        return files
    except FileNotFoundError:
        print(f"Directory '{directory}' not found.")
        return []
    except PermissionError:
        print(f"Permission denied for directory '{directory}'.")
        return []
```

```python
def get_record(video_id):
    try:
       yt = YouTube(f"https://www.youtube.com/watch?v={video_id}")

       return {"video_id": video_id, "publish_date": yt.publish_date, "title": yt.title, "views": yt.views, "length": yt.length}

    except Exception as e:
        print(f'An error occurred: {str(e)}')
```

## Derive records

```python
videoIDs = list_files("../../texts/DialogueWorks/Wolff-Hudson") 
len(videoIDs)
```

```python
get_record(videoIDs[0])
```

```python
records = []
k = 0
for vid in videoIDs:
    try:
        records.append(get_record(vid))
    except:
        print(f"Cannot get record for {vid} at position {k}.")
    k = k + 1

len(records)    
```

```python
records
```

```python
records2 = copy.deepcopy(records)
records2 = [r for r in records if isinstance(r, dict)]
for rec in records2:
    if isinstance(rec, dict):
        rec["publish_date"] = rec["publish_date"].date().isoformat()

records2    
```
